```python
import requests
import pandas as pd
import time

client_id = "45da58455f7f46d5b30b0b8a117fa452"
client_secret = "7dd2efbeaecd4fd08136e682f7610f48"

# 獲取 access_token
def get_access_token():
    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Content-Type": "application/x-www-form-urlencoded"
    }
    data = {
        "grant_type": "client_credentials"
    }
    response = requests.post(url, headers=headers, data=data, auth=(client_id, client_secret))
    return response.json().get('access_token')

# 查詢專輯的音樂類型
def get_album_genre(album_id, access_token):
    url = f"https://api.spotify.com/v1/albums/{album_id}/tracks?market=TW"
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(url, headers=headers)
    data = response.json()

    # 取得專輯中的所有歌曲
    tracks = data.get('items', [])
    genres = set()  # 使用集合避免重複的類型

    # 對於每首歌曲，查詢其詳細資料來獲取音樂類型
    for track in tracks:
        track_id = track.get("id")
        track_genres = get_track_genres(track_id, access_token)
        genres.update(track_genres)  # 更新音樂類型集合

    return list(genres)

# 查詢歌曲的音樂類型
def get_track_genres(track_id, access_token):
    url = f"https://api.spotify.com/v1/audio-features/{track_id}"
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(url, headers=headers)
    data = response.json()
    
    # Spotify API 不直接提供歌曲的 genres，因此這裡返回空列表
    return []  # 根據實際需求來處理音樂類型

# 主程式
def main():
    access_token = get_access_token()

    # 讀取 CSV
    artist_album_ids_path = r"C:\Users\T14 Gen 3\Desktop\KKbox_Spotify_va13\artist_album_ids.csv"
    df = pd.read_csv(artist_album_ids_path)

    # 將查詢結果存入 DataFrame
    result_data = []
    
    for index, row in df.iterrows():
        album_id = row["Artist_album_id"]

        try:
            album_genre_all = get_album_genre(album_id, access_token)

            if album_genre_all:
                result_data.append({"Album_id": album_id, "Album_genre": album_genre_all})
            else:
                print(f"Album ID {album_id} 沒有專輯資料。")
        
            # 添加延遲，防止達到速率限制
            time.sleep(10)  # 每次請求後等待 10 秒
            
        except ValueError as e:
            print(e.args)
    
    result_df = pd.DataFrame(result_data)

    # 保存結果到 CSV
    result_csv_path = r"C:\Users\T14 Gen 3\Desktop\KKbox_Spotify_va13\album_genres.csv"
    result_df.to_csv(result_csv_path, index=False)

if __name__ == "__main__":
    main()
```
