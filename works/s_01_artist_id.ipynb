```python
import requests

import pandas as pd

client_id = "45da58455f7f46d5b30b0b8a117fa452"
client_secret = "7dd2efbeaecd4fd08136e682f7610f48"


# 獲取 access_token
def get_access_token():
    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Content-Type": "application/x-www-form-urlencoded"
    }
    data = {
        "grant_type": "client_credentials"
    }
    response = requests.post(url, headers=headers, data=data, auth=(client_id, client_secret))
    return response.json().get('access_token')


# 查詢歌手 ID
def get_artist_id(artist_name, access_token):
    url = f"https://api.spotify.com/v1/search?q={artist_name}&type=artist"
    headers = {
        "Authorization": f"Bearer {access_token}"
    }
    response = requests.get(url, headers=headers)
    data = response.json()

    # 取得歌手 ID
    items = data.get('artists', {}).get('items', [])
    artist_id = items[0].get('id') if items else None

    return artist_id


# 主程式
def main():
    access_token = get_access_token()

    # 讀取 CSV
    spotify_chart_sample_data_path = r"C:\Users\T14 Gen 3\Desktop\KKbox_Spotify\spotify_chart_sample_data.csv"
    df = pd.read_csv(spotify_chart_sample_data_path)
      
    artist_ids = {"artist": "artist_id"}

    for index, row in df.iterrows():
        artist = row['artist']
        artist_id = get_artist_id(artist, access_token)
        artist_ids[artist] = artist_id
    
    # 將查詢結果存入 DataFrame
    result_df = pd.DataFrame(list(artist_ids.items()), columns=['Artist', 'Artist ID'])

    # 保存結果到 CSV
    result_csv_path = r"C:\Users\T14 Gen 3\Desktop\KKbox_Spotify\artist_ids.csv"
    result_df.to_csv(result_csv_path, index=False)
    

if __name__ == "__main__":
    main()
```
